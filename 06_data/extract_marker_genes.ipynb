```python
import scanpy as sc
```

```python
path = "/project/Wellcome_Discovery/datashare/lturiano/data/"
adata = sc.read_h5ad(path + "rna_filt_aligned.h5ad")
```

```python
sc.tl.rank_genes_groups(
    adata,
    groupby="cell_type",
    method="wilcoxon",      # robust default for scRNA-seq
    use_raw=False,          # set True only if you stored raw in adata.raw
    pts=True                # adds % expressing (useful for marker filtering)
)
```

```python
top_n = 10
marker_dict = {}
for ct in adata.obs["cell_type"].unique():
    df_ct = sc.get.rank_genes_groups_df(adata, group=ct)
    marker_dict[ct] = df_ct["names"].head(top_n).tolist()

marker_dict
```
